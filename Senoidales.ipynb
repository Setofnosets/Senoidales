<h1>Propiedades de las señales senoidales<h1>
<p>El primer módulo correspondiente a la capa física es la fuente de información, esta se refiere a la señal que porta la información enviada. Estas señales pueden ser analógicas o digitales.</p>
<p>La señal principal que se manjea es la señal senoidal, esta es una curva matemática definida en base a la función matemática seno. En el caso de una señal senoidal, esta se define en función del tiempo como:</p>
<p>$A\sin(2\pi ft + \phi)$</p>
<p>Donde $A$ es la amplitud de la onda (La mayor desviación desde 0), $f$ es la frecuencia (el número de oscilaciones en un segundo), $t$ es el tiempo y la variable independiente de la función y $\phi$ corresponde a la fase.</p>
<p>En este notebook se detallan las propiedades de la señal senoidal y como se ve afectada por la modificación de sus parámetros.</p>
<p>A continuación se muestra una señal senoidal con $A = 1$, $f = 1$ y $\phi = 1$ en el rango [0, 2$\pi$]:<p>

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.pyplot import figure
import numpy as np

%matplotlib inline

def sine_wave(A: float, f: float, phi: float, t:float) -> float:
    return A*sin(2*pi*f*t + phi)

amplitud = 1.
frecuencia = 1.
fase = 1.
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitud, frecuencia, fase, tiempo) for tiempo in x]

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-2, 2, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(fase/10))
ax.xaxis.grid(True,'minor',linewidth=0.5)
ax.yaxis.grid(True,'minor',linewidth=0.5)
plt.grid(True, linewidth=1, color='#e5e5e5', linestyle='-')

#Mostrar Grafica
plt.plot(x,y)
plt.show()
```

<h3>Señal periódica</h3>
<p>Una señal o función periódica es una función que repite sus valores en intervalos regulares, al valor de x que toma la señal en repetir su forma se le conoce como periodo</p>
<p>Una señal senoidal es una señal periódica debido a que la señal repite su forma</p>
<p>En el caso de una señal senoidal el periodo se puede definir como $T = \frac{1}{f}$ por lo que podemos representar a la fórmula como $A\sin(\frac{2\pi t}{T} + \phi)$, esto quiere decir que conforme aumenta la frecuencia, el periodo será menor.</p>
<p>En el siguiente ejemplo se muestra un ejemplo $A = 1$, $\phi = 1$ y $T = 1$, para modificar el periodo se debe modificar la variable periodo del programa (Nótese como la señal repite su forma cada $T$ segundos:</p>

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.pyplot import figure
import numpy as np

def sine_wave(A: float, p: float, phi: float, t:float) -> float:
    return A*sin((2*pi*t)/p + phi)

#Variable a modificar
periodo = 1.




amp_max = 0.
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave(1, periodo, 0, tiempo) for tiempo in x]
amp_max = y.index(max(y)) * step

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.xaxis.grid(True,'minor',linewidth=0.5)
ax.yaxis.grid(True,'minor',linewidth=0.5)
plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
plt.annotate("", xy=(amp_max,1), xytext=(amp_max+periodo,1), arrowprops=dict(arrowstyle='|-|'))
plt.annotate("Periodo: "+str(periodo), xy=(amp_max+periodo/2,1), xytext=(amp_max+periodo/2,1+0.1))
#Mostrar Grafica
plt.plot(x,y)
plt.show()
```

<h3>Frecuencia: </h3>
<p>La frecuencia se puede definir como el número de ocurrencias de un evento periódico por unidad de tiempo. En el caso de una señal
senoidal, la frecuencia se puede considerar como el número de repeticiones de su forma en una unidad de tiempo, su inverso es el periodo.</p>
<p>Por ejemplo, para una frecuencia de 5, la señal se repetiría 5 veces por cada valor de $t = x$, esto implica que entre mayor la frecuencia, menor será el periodo</p>
<p>En el siguiente código, la frecuencia se puede modificar mediante la variable frecuencia, observe como conforme se incrementa el valor de la frecuencia, el número de repeticiones por segundo aumenta:</p>

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.pyplot import figure
import numpy as np

def sine_wave(A: float, f: float, phi: float, t:float) -> float:
    return A*sin((2*pi*t*f) + phi)

#Variable a modificar
frecuencia = 1.




x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave(1, frecuencia, 0, tiempo) for tiempo in x]

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.xaxis.grid(True,'minor',linewidth=0.5)
ax.yaxis.grid(True,'minor',linewidth=0.5)
plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
plt.annotate("", xy=(0,1), xytext=(1,1), arrowprops=dict(arrowstyle='|-|'))
plt.annotate("Frecuencia: "+str(frecuencia), xy=(0.5,1), xytext=(0.5,1+0.1))
#Mostrar Grafica
plt.plot(x,y)
plt.show()
```

<h3>Amplitud:</h3>
<p>La amplitud de una señal se puede ver como el valor máximo y mínimo que alcanza una señal periodica en un periodo dado, otra
manera de interpretarlo es como el valor absoluto del cambio medido desde el origen de una señal en un periodo</p>
<p>La amplitud afecta la "altura" máxima que alcanzará la señal, este parámetro es independiente de los demás y no afecta nada más</p>
<p>En el siguiente ejemplo la amplitud se puede modificar por medio de la variable con el mismo nombre, vease como la amplitud afecta
el tamaño de la señal: </p>

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.pyplot import figure
import numpy as np

def sine_wave(A: float, f: float, phi: float, t:float) -> float:
    return A*sin((2*pi*t*f) + phi)

#Variable a modificar
amplitud = 1.



amp_max = 0.
x_max = 2*pi
x_min = 0
step = 0.01
x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitud, 1, 1, tiempo) for tiempo in x]
amp_max = y.index(max(y)) * step

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.xaxis.grid(True,'minor',linewidth=0.5)
ax.yaxis.grid(True,'minor',linewidth=0.5)
plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
plt.annotate("", xy=(amp_max+1/1,0), xytext=(amp_max+1/1,amplitud), arrowprops=dict(arrowstyle='|-|'))
plt.annotate("Amplitud: "+str(amplitud), xy=(amp_max+1/1,amplitud), xytext=(amp_max+1/1+0.1,amplitud-1))

#Mostrar Grafica
plt.plot(x,y)
plt.show()
```

<h3>Fase: </h3>
<p>El concepto de fase se basa en una magnitud cíclica representando el periodo transcurrido desde la medición, esto quiere decir que la fase se conoce como cuando se comienza la captura de la señal, si la fase es 0, la señal se captura desde el origen mientras que si la señal es $\frac{\pi}{2}$ la señal originará desde la cresta (amplitud positiva), los valores de la fase se toman en radianes por lo que su valor máximo es $2\pi$ radianes antes de repetirse</p>
<p>Los valores de fase más relevantes son: $0, \frac{\pi}{2}, \pi, \frac{3\pi}{2}$ y $2\pi$ los cuales representan el origen, la cresta, el segundo origen, el valle y el origen respectivamente ($0$ y $2\pi$ dan resultados idénticos y se repite el patrón)</p>
<p>En el siguiente ejemplo la fase se modifica mediante su variable, observece como el inicio de la señal cambia con diferentes valores de fase (Para utilizar los valores de $\pi$ escribir únicamente "pi":</p>

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.pyplot import figure
import numpy as np

def sine_wave(A: float, f: float, phi: float, t:float) -> float:
    return A*sin((2*pi*t*f) + phi)

#Variable a modificar
fase = 3*pi/2



amp_max = 0.
x_max = 2*pi
x_min = 0
step = 0.01
x = np.arange(x_min, x_max, step)
y = [sine_wave(1, 1, fase, tiempo) for tiempo in x]
amp_max = y.index(max(y)) * step

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.xaxis.grid(True,'minor',linewidth=0.5)
ax.yaxis.grid(True,'minor',linewidth=0.5)
plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
plt.annotate("Fase: " + str(fase), xy=(-0.2,1.1), xytext=(-0.2,1.1))

#Mostrar Grafica
plt.plot(x,y)
plt.show()
```

<h3>Longitud de Onda: </h3>
<p>La longitud de onda es la distancia tras la cual se repite la forma de la señal, esta se puede repetir entre dos puntos cualesquiera en la misma fase.</p>
<p>La longitud de onda se relaciona con el periodo y parecen similar, sin embargo mientras que el periodo se mide en segundos, la longitud de onda se mide en metros. La longitud de onda se representa con el símbolo $\lambda$.</p>
<p>En el siguiente ejemplo la longitud de onda se puede modificar mediande la variable longitud, en este caso entre menor sea la longitud de onda, mayor será la frecuencia:</p>

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.pyplot import figure
import numpy as np


def sine_wave(A: float, p: float, phi: float, t: float) -> float:
    return A * sin((2 * pi * t) / p + phi)


#Variable a modificar
longitud = 3.

amp_half = 0.
x_max = 2 * pi
x_min = 0
step = 0.01


x = np.arange(x_min, x_max, step)
y = [sine_wave(1, longitud, 0, tiempo) for tiempo in x]

for i in range(len(y)):
    if abs(y[i] - 0.5) < step:
        amp_half = i * step
        break

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.xaxis.grid(True, 'minor', linewidth=0.5)
ax.yaxis.grid(True, 'minor', linewidth=0.5)
plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
plt.annotate("", xy=(amp_half, 0.5), xytext=(amp_half + longitud, 0.5), arrowprops=dict(arrowstyle='|-|'))
plt.annotate("λ: " + str(longitud), xy=(amp_half + longitud / 2, 0.6), xytext=(amp_half + longitud / 2, 0.6))
#Mostrar Grafica
plt.plot(x, y)
plt.show()
```
