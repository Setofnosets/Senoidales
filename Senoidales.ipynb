<p>El primer módulo correspondiente a la capa física es la fuente de información, esta se refiere a la señal que porta la información enviada. Estas señales pueden ser analógicas o digitales.</p>
<p>La señal principal que se manjea es la señal senoidal, esta es una curva matemática definida en base a la función matemática seno. En el caso de una señal senoidal, esta se define en función del tiempo como:</p>
<p>$A\sin(2\pi ft + \phi)$</p>
<p>Donde $A$ es la amplitud de la onda (La mayor desviación desde 0), $f$ es la frecuencia (el número de oscilaciones en un segundo), $t$ es el tiempo y la variable independiente de la función y $\phi$ corresponde a la fase.</p>
<p>En este notebook se detallan las propiedades de la señal senoidal y como se ve afectada por la modificación de sus parámetros.</p>
<p>A continuación se muestra una señal senoidal con $A = 1$, $f = 1$ y $\phi = 1$ en el rango [0, 2$\pi$]:<p>

```python
from math import sin, pi, cos
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.pyplot import figure
import numpy as np


# %matplotlib inline

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A * sin((2 * pi * t * f) + phi)


def mensaje(M: float, f: float, phi: float, t: float) -> float:
    return M * sin((2 * pi * t * f) + phi)


def modulacion_amplitud(A: float, M: float, fc: float, fm: float, phi: float, t: float) -> float:
    indice = M / A
    return (1 + mensaje(indice, fm, phi, t)) * sine_wave(A, fc, phi, t)


def mostrar_grafica(x: None, y: list, title: str) -> None:
    ax = plt.figure().add_axes([0, 0, 1, 1])
    ax.set_title(title)
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    if x is not None:
        plt.plot(x, y, 'g')
    else:
        plt.plot(y, 'g')

amplitud = 1.
frecuencia = 1.
fase = 1.
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitud, frecuencia, fase, tiempo) for tiempo in x]

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-2, 2, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(fase/10))
ax.xaxis.grid(True,'minor',linewidth=0.5)
ax.yaxis.grid(True,'minor',linewidth=0.5)
plt.grid(True, linewidth=1, color='#e5e5e5', linestyle='-')

#Mostrar Grafica
plt.plot(x,y)
plt.show()
```

<h3>Señal periódica</h3>
<p>Una señal o función periódica es una función que repite sus valores en intervalos regulares, al valor de x que toma la señal en repetir su forma se le conoce como periodo</p>
<p>Una señal senoidal es una señal periódica debido a que la señal repite su forma</p>
<p>En el caso de una señal senoidal el periodo se puede definir como $T = \frac{1}{f}$ por lo que podemos representar a la fórmula como $A\sin(\frac{2\pi t}{T} + \phi)$, esto quiere decir que conforme aumenta la frecuencia, el periodo será menor.</p>
<p>En el siguiente ejemplo se muestra un ejemplo $A = 1$, $\phi = 1$ y $T = 1$, para modificar el periodo se debe modificar la variable periodo del programa (Nótese como la señal repite su forma cada $T$ segundos:</p>

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.pyplot import figure
import numpy as np

%matplotlib inline

def sine_wave_periodic(A: float, p: float, phi: float, t:float) -> float:
    return A*sin((2*pi*t)/p + phi)

#Variable a modificar
periodo = 1.




amp_max = 0.
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave_periodic(1, periodo, 0, tiempo) for tiempo in x]
amp_max = y.index(max(y)) * step

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.xaxis.grid(True,'minor',linewidth=0.5)
ax.yaxis.grid(True,'minor',linewidth=0.5)
plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
plt.annotate("", xy=(amp_max,1), xytext=(amp_max+periodo,1), arrowprops=dict(arrowstyle='|-|'))
plt.annotate("Periodo: "+str(periodo), xy=(amp_max+periodo/2,1), xytext=(amp_max+periodo/2,1+0.1))
#Mostrar Grafica
plt.plot(x,y)
plt.show()
```

<h3>Frecuencia: </h3>
<p>La frecuencia se puede definir como el número de ocurrencias de un evento periódico por unidad de tiempo. En el caso de una señal
senoidal, la frecuencia se puede considerar como el número de repeticiones de su forma en una unidad de tiempo, su inverso es el periodo.</p>
<p>Por ejemplo, para una frecuencia de 5, la señal se repetiría 5 veces por cada valor de $t = x$, esto implica que entre mayor la frecuencia, menor será el periodo</p>
<p>En el siguiente código, la frecuencia se puede modificar mediante la variable frecuencia, observe como conforme se incrementa el valor de la frecuencia, el número de repeticiones por segundo aumenta:</p>

```python


#Variable a modificar
frecuencia = 1.




x_min = 0
step = 0.01

x = np.arange(x_min, 1, step)
y = [sine_wave(1, frecuencia, 0, tiempo) for tiempo in x]

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.xaxis.grid(True,'minor',linewidth=0.5)
ax.yaxis.grid(True,'minor',linewidth=0.5)
plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
plt.annotate("Frecuencia: "+str(frecuencia), xy=(0.5,1), xytext=(0.5,1+0.1))
#Mostrar Grafica
plt.plot(x,y)
plt.show()
```

<h3>Amplitud:</h3>
<p>La amplitud de una señal se puede ver como el valor máximo y mínimo que alcanza una señal periodica en un periodo dado, otra
manera de interpretarlo es como el valor absoluto del cambio medido desde el origen de una señal en un periodo</p>
<p>La amplitud afecta la "altura" máxima que alcanzará la señal, este parámetro es independiente de los demás y no afecta nada más</p>
<p>En el siguiente ejemplo la amplitud se puede modificar por medio de la variable con el mismo nombre, vease como la amplitud afecta
el tamaño de la señal: </p>

```python


#Variable a modificar
amplitud = 1.



amp_max = 0.
x_max = 2*pi
x_min = 0
step = 0.01
x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitud, 1, 1, tiempo) for tiempo in x]
amp_max = y.index(max(y)) * step

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.xaxis.grid(True,'minor',linewidth=0.5)
ax.yaxis.grid(True,'minor',linewidth=0.5)
plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
plt.annotate("", xy=(amp_max+1/1,0), xytext=(amp_max+1/1,amplitud), arrowprops=dict(arrowstyle='|-|'))
plt.annotate("Amplitud: "+str(amplitud), xy=(amp_max+1/1,amplitud), xytext=(amp_max+1/1+0.1,amplitud-1))

#Mostrar Grafica
plt.plot(x,y)
plt.show()
```

<h3>Fase: </h3>
<p>El concepto de fase se basa en una magnitud cíclica representando el periodo transcurrido desde la medición, esto quiere decir que la fase se conoce como cuando se comienza la captura de la señal, si la fase es 0, la señal se captura desde el origen mientras que si la señal es $\frac{\pi}{2}$ la señal originará desde la cresta (amplitud positiva), los valores de la fase se toman en radianes por lo que su valor máximo es $2\pi$ radianes antes de repetirse</p>
<p>Los valores de fase más relevantes son: $0, \frac{\pi}{2}, \pi, \frac{3\pi}{2}$ y $2\pi$ los cuales representan el origen, la cresta, el segundo origen, el valle y el origen respectivamente ($0$ y $2\pi$ dan resultados idénticos y se repite el patrón)</p>
<p>En el siguiente ejemplo la fase se modifica mediante su variable, observece como el inicio de la señal cambia con diferentes valores de fase (Para utilizar los valores de $\pi$ escribir únicamente "pi":</p>

```python

def fases (x: float, y: float, titulo: str) -> None:
    ax = plt.figure().add_subplot(1, 1, 1)
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
    ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
    ax.xaxis.set_minor_locator(MultipleLocator(0.1))
    ax.yaxis.set_minor_locator(MultipleLocator(0.1))
    ax.xaxis.grid(True,'minor',linewidth=0.5)
    ax.yaxis.grid(True,'minor',linewidth=0.5)
    plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
    plt.annotate("Fase: " + titulo, xy=(-0.2,1.1), xytext=(-0.2,1.1))
    plt.plot(x, y, 'g')


#Variable a modificar
fase = 0

amp_max = 0.
x_max = 2*pi
x_min = 0
step = 0.01
x = np.arange(x_min, x_max, step)
y = [sine_wave(1, 1, fase, tiempo) for tiempo in x]

#Mostrar Grafica
fig = plt.figure()
fases(x, y, "0")
fase = pi/2
y = [sine_wave(1, 1, fase, tiempo) for tiempo in x]
fig = plt.figure()
fases(x, y, "π/2")
fase = pi
y = [sine_wave(1, 1, fase, tiempo) for tiempo in x]
fig = plt.figure()
fases(x, y, "π")
fase = 2*pi/3
y = [sine_wave(1, 1, fase, tiempo) for tiempo in x]
fig = plt.figure()
fases(x, y, "2π/3")
plt.show()
```

<h3>Longitud de Onda: </h3>
<p>La longitud de onda es la distancia tras la cual se repite la forma de la señal, esta se puede repetir entre dos puntos cualesquiera en la misma fase.</p>
<p>La longitud de onda se relaciona con el periodo y parecen similar, sin embargo mientras que el periodo se mide en segundos, la longitud de onda se mide en metros. La longitud de onda se representa con el símbolo $\lambda$.</p>
<p>En el siguiente ejemplo la longitud de onda se puede modificar mediande la variable longitud, en este caso entre menor sea la longitud de onda, mayor será la frecuencia:</p>

```python

#Variable a modificar
longitud = 3.

amp_half = 0.
x_max = 2 * pi
x_min = 0
step = 0.01


x = np.arange(x_min, x_max, step)
y = [sine_wave_periodic(1, longitud, 0, tiempo) for tiempo in x]

for i in range(len(y)):
    if abs(y[i] - 0.5) < step:
        amp_half = i * step
        break

#Agregar Ejes
ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.xaxis.grid(True, 'minor', linewidth=0.5)
ax.yaxis.grid(True, 'minor', linewidth=0.5)
plt.grid(True, linewidth=0.5, color='#e5e5e5', linestyle='-')
plt.annotate("", xy=(amp_half, 0.5), xytext=(amp_half + longitud, 0.5), arrowprops=dict(arrowstyle='|-|'))
plt.annotate("λ: " + str(longitud), xy=(amp_half + longitud / 2, 0.6), xytext=(amp_half + longitud / 2, 0.6))
#Mostrar Grafica
plt.plot(x, y)
plt.show()
```

<h3>Señal compuesta:</h3>
<p>A una señal se le conoce como simple cuando consiste de una sola onda senoidal que no se puede descomponer de ninguna manera. Sin embargo, en la vida real la mayoría de las señales utilizadas se conocen como compuestas porque estas señales se pueden descomponer en unconjunto de ondas senoidales simples.</p>
<p>En el siguiente ejemplo se muestra una señal compuesta: </p>

```python

def mostrar_grafica2(x: list, y: list, title: str, fig: any, y1: list=None, y2: list=None) -> None:
    fig.suptitle(title, fontsize=16)

    ax = fig.add_axes([0,0,1,1])
    #ax.set_title(title)
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
    ax.yaxis.set_ticks(np.arange(-100, 100, 0.5))
    ax.xaxis.set_minor_locator(MultipleLocator(0.1))
    ax.yaxis.set_minor_locator(MultipleLocator(0.1))
    ax.xaxis.grid(True, 'minor', linewidth=0.5)
    ax.yaxis.grid(True, 'minor', linewidth=0.5)
    ax.plot(x, y, 'g')
    if y1 is not None:
        ax.plot(x, y1, 'r')
    if y2 is not None:
        ax.plot(x, y2, 'b')


amp_max = 0.
x_max = 2*pi
x_min = 0
step = 0.01
x = np.arange(x_min, x_max, step)
#Señal con A = 1, f = 4 y fase = 0
signal1 = [sine_wave(1,4,0,tiempo) for tiempo in x]
#Señal con A = 2.5, f = 0.5 y fase = 0
signal2 = [sine_wave(2.5,0.5,0,tiempo) for tiempo in x]
#Señal con A = 3, f = 1.5 y fase = 0
signal3 = [sine_wave(3,1.5,0,tiempo) for tiempo in x]

fig = plt.figure()
mostrar_grafica2(x, signal1, "Señal 1", fig)
fig = plt.figure()
mostrar_grafica2(x, signal2, "Señal 2", fig)
fig = plt.figure()
mostrar_grafica2(x, signal3, "Señal 3", fig)

#Mostrar Graficas en una sola
fig = plt.figure()
mostrar_grafica2(x, signal1, "Señales", fig, signal2, signal3)
fig = plt.figure()
compuesta = [signal1[i] + signal2[i] + signal3[i] for i in range(len(signal1))]
mostrar_grafica2(x, compuesta, "Señal compuesta", fig)

plt.show()
```

<h2>Modulación por amplitud:</h2>
<p>Un método de modulación utilizado frecuentemente es la modulación por amplitud la cual modifica la amplitud de una señal en proporción a la información enviada. Esta modulación consiste de dos señales simples: una señal de información y una señal de transmisión. La señal de información es una onda con amplitud variable contiene la información en sí del mensaje mientras que la señal de transmisión es una señal con una frecuencia determinada.</p>
<p>Al realizar el proceso de modulación por amplitud se genera una sepñal compuesta por la frecuencia de la señal de transmición y la amplitud se ve modificada de manera similar a la señal de información. La señal resultante se puede representar de la siguiente manera: $y(t) = [1+m\cos(2\pi f$<sub>M</sub>$t+\phi)]A\sin(2\pi f$<sub>C</sub>$t$ Donde $m$ es el índice de modulación, $f$<sub>M</sub> es la frecuencia del mensaje y $f$<sub>C</sub> es la frecuencia de la señal de transmisión. En el siguiente código se muestra un ejemplo de modulación por amplitud:</p>

```python


x = np.linspace(0,1,1000)
#Señal con A = 100, f = 2 y fase = 0
mensaje1 = [mensaje(100,2,0,tiempo) for tiempo in x]
#Señal con A = 10, f = 100 y fase = 0
transmision = [sine_wave(10,100,0,tiempo) for tiempo in x]
#Señal modulada
modulada = [modulacion_amplitud(10,100,100,2,0,tiempo) for tiempo in x]

fig = plt.figure()
mostrar_grafica(x, mensaje1, "Mensaje")
fig = plt.figure()
mostrar_grafica(x, transmision, "Señal de transmision")

#Mostrar Graficas moduladas
fig = plt.figure()
mostrar_grafica(x, modulada, "Señal modulada")
plt.show()
```

<h3>Ancho de banda</h3>
<p>El ancho de banda genaralmente se reconoce como la máxima velocidad de transferencia de datos de una señal. En comunicaciones el ancho de banda se refiere a la diferencia entre las frecuencias más altas y más bajas de las partes que constituyen la señal, si una señal es de 2 Hertz y otra de 1 Hertz, entonces el ancho de banda será de 1 Hertz. Otra manera de pensarlo es como el rango de frecuencias de una señal, por ejemplo en el código siguiente se mostrará un ancho de banda de $4000 - 400 = 3600$ Hz</p>
<h3>Transformada de Fourier</h3>
<p>La transformada de Fourier es un método matemático inventando for Joseph Fourier que permite descomponer una función compuesta en sus componentes, este mecanismo es el que nos permite demodular una señal y se encuentra definida por la siguiente fórmula: $\int_{-\infty}^{\infty} f(x) e^{-i2\pi \xi x} dx, $</p>
<h3>Espectro de frecuencias</h3>
<p>El espectro de frecuencias simplemente se refiere a la distribución de amplitudes para cada frecuencia, es decir, en este caso son los resultados producidos por la transformada de Fourier donde se muestran las frecuencias presentes en la señal dada. Lo anterior significa que es posible filtrar ciertas frecuencias para despues realizar una transformada de Fourier inversa y así obtener una de nuestras señales originales.</p>
<p>El siguiente ejemplo muestra el resultado de una transformada de fourier de una señal modulada similar a la usada anteriormente mediante la funcion fft de numpy, vease como se muestra el espectro de frecuencias de ambas señales que conforman la señal modulada en sus partes real(positivo) e imaginaria(negativo):</p>

```python

x = np.linspace(0,1,1000)

#Señal modulada
modulada = [modulacion_amplitud(2,30,4000,400,0,tiempo) for tiempo in x]

yf = np.fft.fft(modulada)
xf = np.fft.fftfreq(len(yf), 0.001)

#Mostrar Graficas moduladas
fig = plt.figure()
mostrar_grafica(x, modulada, "Señal modulada")
fig = plt.figure()
#Transformada de Fourier
plt.suptitle("Espectro de frecuencias", fontsize=16)
plt.plot(xf, np.abs(yf), 'g')

plt.show()
```
